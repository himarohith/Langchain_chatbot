```python
import os
from docx import Document

def clean_docx_file(input_file, output_file):
    doc = Document(input_file)
    
    for paragraph in doc.paragraphs:
        # Replace multiple spaces with a single space
        paragraph.text = ' '.join(paragraph.text.split())
    
    doc.save(output_file)

def clean_all_docx_files_in_folder(folder_path):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.docx'):
                input_file = os.path.join(root, file)
                output_file = os.path.join(root, 'cleaned_' + file)
                clean_docx_file(input_file, output_file)
                print(f"Spaces cleaned in {input_file} and saved to {output_file}")

if __name__ == '__main__':
    folder_path = 'C:/Users/vthammuneni/Desktop/Hima_Chatbot/pdf_word/'  # Replace with the path to your folder containing DOCX files
    clean_all_docx_files_in_folder(folder_path)
```
